```python
import re
import string
import pandas as pd
from math import log
import os
import random
import re
import string
from collections import Counter
from sklearn.cluster import KMeans

import nltk
import numpy as np
import pandas as pd

from gensim.models import Word2Vec

from nltk import word_tokenize
from nltk.tokenize import wordpunct_tokenize
from nltk.corpus import stopwords

import warnings
warnings.filterwarnings("ignore")
nltk.download("stopwords")


SEED = 42
random.seed(SEED)
os.environ["PYTHONHASHSEED"] = str(SEED)
np.random.seed(SEED)
```

```python
clusters = 2
def mbkmeans_clusters(X, k):
    km = KMeans(n_clusters=k).fit(X)
    return km.labels_
```

```python
final_counts = []
```

```python
def read_and_store(path):
    df = pd.read_pickle(path)
    df.isna().sum()
    df = df.loc[ df.histograms.map(lambda x: ~np.isnan(x).any()), :]
    cluster_labels = mbkmeans_clusters(
        X=list(df['histograms'].values),
        k=clusters,
    )
     
    final_counts.append(Counter(cluster_labels))
    
    
    
    
```
